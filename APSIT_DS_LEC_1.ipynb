```python
class Stack_Example:
    def __init__(self,size):
        self.tos=-1
        self.slist=[]
        for i in range(size):
            self.slist.append(0)
        self.maxsize=size
    def push(self,element):
        self.tos+=1
        self.slist[self.tos]=element
        print("pushed:",element)
    def pop(self):
        temp=self.slist[self.tos]
        self.tos-=1
        return(temp)
    def isfull(self):
        if self.tos==self.maxsize-1:
            return True
        else:
            return False
    def isempty(self):
        if self.tos==-1:
            return True
        else:
            return False
    def peek(self):
        return(self.slist[self.tos])
    def print_stack(self):
        for i in range(self.tos,-1,-1):
            print(self.slist[i])
```

```python
#menudriven code for stack
'''start with size of stack----user
1.push 2.pop 3.peek 4.print 0.Exit
before push check full or not
before pop check empty of not
before peek check empty of not
before print check empty of not
also check for wrong input
'''

size=int(input("Enter size:"))
obj=Stack_Example(size)
while True:
    ch=int(input("\n1.Push\n2.Pop\n3.Peek\n4.Print\n0.Exit\n:"))
    if ch==1:
        if obj.isfull()!=True:
            e=int(input("Enter data:"))
            obj.push(e)
        else:
            print("stack is full")
    elif ch==2:
        if obj.isempty()!=True:
            e=obj.pop()
            print(e,"poped")
        else:
            print("stack is empty")
    elif ch==3:
        if obj.isempty()!=True:
            e=obj.peek()
            print(e,"at peek")
        else:
            print("stack is empty")
    elif ch==4:
        if obj.isempty()!=True:
            print("Stack has:")
            obj.print_stack()
        else:
            print("stack is empty")
    elif ch==0:
         print("THANKS FOR USING THE CODE:")
         break
    else:
        print("Wrong option given")
```

```python
class Queue_Example:
    def __init__(self,size):
        self.front,self.rear=0,-1
        self.qlist=[]
        for i in range(size):
            self.qlist.append(0)
        self.maxsize=size
    def enqueue(self,element):
        self.rear+=1
        self.qlist[self.rear]=element
        print("Enqueued:",element)
    def dequeue(self):
        temp=self.qlist[self.front]
        self.front+=1
        return(temp)
    def isfull(self):
        if self.rear==self.maxsize-1:
            return True
        else:
            return False
    def isempty(self):
        if self.rear<self.front:
            return True
        else:
            return False
    def print_queue(self):
        for i in range(self.front,self.rear+1):
            print(self.qlist[i],end="-")
```

```python
class Circular_Queue_Example:
    def __init__(self,size):
        self.front,self.rear=0,-1
        self.count=0
        self.qlist=[]
        for i in range(size):
            self.qlist.append(0)
        self.maxsize=size
    def enqueue(self,element):
        self.rear=(self.rear+1)%self.maxsize
        self.count+=1
        self.qlist[self.rear]=element
        print("Enqueued:",element)
    def dequeue(self):
        temp=self.qlist[self.front]
        self.front=(self.front+1)%self.maxsize
        self.count-=1
        return(temp)
    def isfull(self):
        if self.count==self.maxsize:
            return True
        else:
            return False
    def isempty(self):
        if self.count==0:
            return True
        else:
            return False
    def print_queue(self):
        c=0
        i=self.front
        while c<self.count:
            print(self.qlist[i],end="-")
            c+=1
            i=(i+1)%self.maxsize

size=int(input("Enter size:"))
obj=Circular_Queue_Example(size)
while True:
    ch=int(input("\n1.Enqueue\n2.Dequeue\n3.Print\n0.Exit\n:"))
    if ch==1:
        if obj.isfull()!=True:
            e=int(input("Enter data:"))
            obj.enqueue(e)
        else:
            print("queue is full")
    elif ch==2:
        if obj.isempty()!=True:
            e=obj.dequeue()
            print(e,"dequeued")
        else:
            print("queue is empty")
    elif ch==3:
        if obj.isempty()!=True:
            print("Queue has:")
            obj.print_queue()
        else:
            print("Queue is empty")
    elif ch==0:
         print("THANKS FOR USING THE CODE:")
         break
    else:
        print("Wrong option given")
```
